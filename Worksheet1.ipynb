```python
#task 3
lenna_grey=lenna_img.convert("L")
lenna_grey_arr=np.array(lenna_grey)
height,width=lenna_grey_arr.shape
data = lenna_grey_arr . reshape ( height , width )
print(data)
display(lenna_grey)
```

```python
mean = np.mean(data,axis = 0)
centered_data = data - mean
print(centered_data)
```

```python
# Compute covariance matrix
cov_matrix = np.cov( centered_data , rowvar = False )
```

```python
eigenvalues, eigenvectors = np.linalg.eigh(cov_matrix)

#sorting in descending order

sorted_indices = np.argsort(eigenvalues)[::-1]
eigenvalues = eigenvalues[sorted_indices]
eigenvectors = eigenvectors[:, sorted_indices]
```

```python
explained_variance_ratio = eigenvalues / np.sum(eigenvalues)
cumulative_variance_ratio = np.cumsum(explained_variance_ratio)
plt.plot(cumulative_variance_ratio)
plt.xlabel('Number of Principal Components')
plt.ylabel('Cumulative Explained Variance Ratio')
plt.grid(True)
plt.show()
```

```python
a=10
b=20
c=50
d=100
e=150
components_a = eigenvectors[:,:a]
components_b = eigenvectors[:,:b]
components_c = eigenvectors[:,:c]
components_d = eigenvectors[:,:d]
components_e = eigenvectors[:,:e]
```

```python
#compressing the image
compressed_data_a = np.dot(centered_data,components_a)
compressed_data_b = np.dot(centered_data,components_b)
compressed_data_c = np.dot(centered_data,components_c)
compressed_data_d = np.dot(centered_data,components_d)
compressed_data_e = np.dot(centered_data,components_e)
print(centered_data.shape)
print(components_a.shape)
print(compressed_data_a.shape)
```

```python
#decompressing the image
decompressed_data_a = np.dot(compressed_data_a,components_a.T) + mean
decompressed_data_b = np.dot(compressed_data_b,components_b.T) + mean
decompressed_data_c = np.dot(compressed_data_c,components_c.T) + mean
decompressed_data_d = np.dot(compressed_data_d,components_d.T) + mean
decompressed_data_e = np.dot(compressed_data_e,components_e.T) + mean
```

```python
plt . figure ( figsize =(12 , 6) )
# Original Image
plt . subplot ( 2 , 3 , 1 )
plt . imshow ( lenna_grey , cmap ='gray' )
plt . title ( 'Original Image' )
plt . axis ( 'off' )
#Deompressed image
plt . subplot ( 2 , 3 , 2 )
plt . imshow ( decompressed_data_a , cmap ='gray' )
plt . title ( 'Decmpressed Image (a=10)' )
plt . axis ( 'off' )

plt . subplot ( 2 , 3 , 3 )
plt . imshow ( decompressed_data_b , cmap ='gray' )
plt . title ( 'Decmpressed Image (b=20)' )
plt . axis ( 'off' )

plt . subplot ( 2 , 3 , 4 )
plt . imshow ( decompressed_data_c , cmap ='gray' )
plt . title ( 'Decmpressed Image (c=50)' )
plt . axis ( 'off' )

plt . subplot ( 2 , 3 , 5 )
plt . imshow ( decompressed_data_d , cmap ='gray' )
plt . title ( 'Decmpressed Image (d=100)' )
plt . axis ( 'off' )

plt . subplot ( 2 , 3 , 6 )
plt . imshow ( decompressed_data_e , cmap ='gray' )
plt . title ( 'Decompressed Image (e=150)' )
plt . axis ( 'off' )
plt . show ()


```
